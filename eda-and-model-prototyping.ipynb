```python
from datetime import datetime

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler


import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams.update({'font.size': 15})
plt.style.use('bmh')
sns.set_palette('colorblind')
```

it looks like all the weather columns are boolean values, which need to be converted to numerical values

before we do that, let's make a copy of the dataset to manipulate. That way we're not making changes in place

```python
f1_data_copy = f1_data.copy()
```

```python
weather_cols = ['weather_warm', 'weather_cold', 'weather_dry',
                'weather_wet', 'weather_cloudy']

for cat in weather_cols:
    f1_data_copy[cat] = f1_data_copy[cat].apply(lambda x: 1 if x == True else 0)
```

```python
# checking to see if the lambda function worked properly
f1_data_copy.head()
```

```python
for col in weather_cols:
    print(f1_data_copy[col].value_counts())
```

Looks like it worked perfectly

let's take a look at a correlation heatmap...

```python
plt.figure(figsize=(16,10), facecolor='w')

heatmap = sns.heatmap(f1_data_copy.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Corelation Heatmap');
```

`grid` - `constructor_standings_pos` are all moderately to strongly related to each other.


it would make sense to do some feature reduction

let's first standardize all of these columns 

```python
cols_to_standardize = ['grid', 'podium', 'driver_points', 
                      'driver_wins', 'constructor_wins', 
                       'constructor_standings_pos']

scaler = StandardScaler()

scaled_cols = scaler.fit_transform(f1_data_copy[cols_to_standardize])
```

```python
for col in range(scaled_cols.shape[1]):
    print(col)
```

```python
# changing the features in the copied df to their scaled versions

for col, scaled_col in zip(cols_to_standardize, range(scaled_cols.shape[1])):
    f1_data_copy[col] = scaled_cols[:, scaled_col]
```
